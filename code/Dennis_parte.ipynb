Apartado c)

\begin{align} 
\hat{\beta} &= \left( \sum_{i=1}^n X_i X_i' + \lambda I_K \right)^{-1} \sum_{i=1}^n X_i Y_i.
\end{align}

Sustituyendo $Y_i = X_i \beta + e_i$ en la siguiente ecuación:
\begin{align}
\hat{\beta} &= \left( \sum_{i=1}^n X_i X_i' + \lambda I_K \right)^{-1} \sum_{i=1}^n X_i \left( X_i \beta + e_i \right), \\
\hat{\beta} &= \left( \sum_{i=1}^n X_i X_i' + \lambda I_K \right)^{-1} \left( \sum_{i=1}^n X_i X_i' \beta + \sum_{i=1}^n X_i e_i \right).
\end{align}

Factorizando $n$:
\begin{align}
\hat{\beta} &= \frac{1}{n} \left( \sum_{i=1}^n X_i X_i' + \frac{\lambda}{n} I_K \right)^{-1} 
\left( \sum_{i=1}^n X_i X_i' \beta + \sum_{i=1}^n X_i e_i \right).
\end{align}

Reescribiendo en términos de promedios:
\begin{align}
\hat{\beta} &= \left( \frac{1}{n} \sum_{i=1}^n X_i X_i' + \frac{\lambda}{n} I_K \right)^{-1}

\end{align}

Tomando el límite cuando $n \to \infty$ y $\frac{\lambda}{n} \to 0$:
\begin{align}
\hat{\beta} &\to \left( X'X \right)^{-1} X'X \beta = \beta.
\end{align}

Por lo tanto, se tiene que:
$\hat{\beta} \to \beta.$

Apartado e)

1. **Consistencia del estimador:**
   Un estimador es consistente si converge en probabilidad al parámetro verdadero conforme $n \to \infty$:
   $$\hat{\beta} \xrightarrow{p} \beta.$$

2. **Varianza condicional:**
   La varianza condicional de un estimador está dada por:
   $$\text{Var}(\hat{\beta} \mid X) = \mathbb{E} \left[ (\hat{\beta} - \mathbb{E}[\hat{\beta} \mid X]) 
   (\hat{\beta} - \mathbb{E}[\hat{\beta} \mid X])' \mid X \right].$$

3. **Supuesto de homocedasticidad:**
   Se asume que $\mathbb{E}[e e' \mid X] = \sigma^2 I$, es decir, los errores son independientes y tienen varianza constante.

4. **Propiedades de matrices:**
   Se utiliza la propiedad de matrices invertibles para manejar la estructura de la varianza.

La varianza condicional de $\hat{\beta}$ dado $X$ está definida como:
\begin{align}
\text{Var}(\hat{\beta} \mid X) &= \mathbb{E} \left[ (\hat{\beta} - \mathbb{E}[\hat{\beta} \mid X]) 
(\hat{\beta} - \mathbb{E}[\hat{\beta} \mid X])' \mid X \right].
\end{align}

Dado el estimador:
\begin{align}
\hat{\beta} &= \left( \sum_{i=1}^n X_i X_i' + \lambda I_K \right)^{-1} \sum_{i=1}^n X_i Y_i,
\end{align}
entonces:
\begin{align}
\hat{\beta} - \mathbb{E}[\hat{\beta} \mid X] &= \left( \sum_{i=1}^n X_i X_i' + \lambda I_K \right)^{-1} 
\left( \sum_{i=1}^n X_i Y_i \right) - \beta \quad + \lambda \left( \sum_{i=1}^n X_i X_i' + \lambda I_K \right)^{-1} \beta.
\end{align}

Expresando matricialmente la varianza:
\begin{align}
\text{Var}(\hat{\beta} \mid X) &= \mathbb{E} \left[ \left( X'X + \lambda I_K \right)^{-1} X'e 
\left( \left( X'X + \lambda I_K \right)^{-1} X'e \right)' \mid X \right].
\end{align}

Utilizando las propiedades de la esperanza y la varianza de errores:
\begin{align}
\text{Var}(\hat{\beta} \mid X) &= \left( X'X + \lambda I_K \right)^{-1} X' \mathbb{E}[e e' \mid X] X 
\left( X'X + \lambda I_K \right)^{-1}.
\end{align}

Asumiendo homocedasticidad y que $\mathbb{E}[e e' \mid X] = \sigma^2 I$:
\begin{align}
\text{Var}(\hat{\beta} \mid X) &= \left( X'X + \lambda I_K \right)^{-1} X'DX 
\left( X'X + \lambda I_K \right)^{-1},
\end{align}
donde $D$ es una matriz diagonal que captura las varianzas de los errores.
